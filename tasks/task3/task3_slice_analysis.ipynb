```python
#导入训练数据

train = pd.read_hdf('./data/train.h5',key='data')
test = pd.read_hdf('./data/test.h5',key='data')

train.reset_index(inplace=True)
test.reset_index(inplace=True)

print('loading success')
```

```python
ec_year=pd.concat([train.iloc[:, 0:11], test.iloc[:, 0:11]],axis=0)
```

```python
year_group = ec_year.groupby(ec_year.date_integraged.dt.year)
```

```python
int_table = pd.DataFrame()

counter = 0
for g in year_group:
    res_year =pd.DataFrame(g[1].ec_number.value_counts()).sort_index()
    res_year['appears']= res_year.ec_number
    res_year.ec_number = res_year.index
    res_year['year']= g[0]
    res_year.reset_index(drop=True, inplace=True)
    if counter ==0:
        int_table = res_year
    else:
        int_table = pd.concat([int_table, res_year], axis=0)
    counter +=1
#     print(g[0])
#     print( res_year)
```
