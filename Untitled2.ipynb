```python
import numpy as np
import pandas as pd
import random
import sys
import os
from tqdm import tqdm
sys.path.append("./tools/")
import funclib
import time

%load_ext autoreload
```

```python
ecdata = pd.read_hdf('./data/emzyme_noemzyme_uncertain.h5',key='data')
ecdata.reset_index(inplace=True)
ecdata= ecdata.iloc[:,1:11]
print('loading success')
```

```python
ecdata
```

```python
year_group = ecdata.groupby(ecdata.date_integraged.dt.year)
```

```python
int_table = pd.DataFrame()

counter = 0
for g in year_group:
    res_year =pd.DataFrame(g[1].ec_number.value_counts()).sort_index()
    res_year['appears']= res_year.ec_number
    res_year.ec_number = res_year.index
    res_year['year']= g[0]
    res_year.reset_index(drop=True, inplace=True)
    if counter ==0:
        int_table = res_year
    else:
        int_table = pd.concat([int_table, res_year], axis=0)
    counter +=1
ec_numbers = sorted(set(int_table.ec_number))
```

```python
max(int_table.appears)
```

```python
show_table = pd.DataFrame(np.zeros([len(ec_numbers), max(int_table.year) - min(int_table.year)+1]), columns=range(min(int_table.year),(max(int_table.year)+1)))
show_table.index = ec_numbers
```

```python
show_table.loc['-', 1986] = 1
```

```python
for index, row in int_table.iterrows():
    show_table.loc[row.ec_number, row.year] = row.appears
```
